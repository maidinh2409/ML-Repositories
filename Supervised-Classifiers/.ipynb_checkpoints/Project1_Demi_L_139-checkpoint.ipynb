#### 1.2.b. A short description, including the targeted values (classes) and features.

This is a dataset from the medical industry that contains data of patients diagnosed with HIV/AIDS who are participating in a study with the goal of observing whether or not each patient died within a certain observation time. The purpose is to examine the performance of two different types of AIDS treatments (with ZDV treatments  and no ZDV treatments)

* Target values: cid - censoring indicator (1 = Death (Patient died within observation time), 0 = Censored (Patient did not die within observation time). **Note: Censored does not mean survival after observation time but can be due to external factors such as the patient stopping the study halfway, accidents, etc. Therefore, outliers may still exist.**
* Features overview:
  + Time (observation time)
  + Personal information (age, weight, race, gender, sexual activity)
  + Medical history (hemophilia (blood clotting disorder), history of IV drug use (drugs injected directly into the vein), Karnofsky score (health index (90-100 meaning normal)))
  + Treatment history (whether or not the patient has been treated with ZDV before)
  + Lab results: counts of two types of immune cells during treatment:
    - CD4: Helper T cells -> The lower the CD4 count, the greater the immune deficiency.
    - CD8: Cytotoxic T cells -> In HIV-infected individuals, CD4 count decreases, and CD8 can increase as a compensatory response to the CD4 deficiency. This results in the CD4:CD8 ratio being potentially low.

### 2.2. Data Cleaning

There are 2 biggest data issues in this dataset:
* Outliers in continuous features: time, cd40, cd420, cd80, cd820, age, weight, preanti are found
* Some column names do not have clear meaning

**<u><span style="color:gray;">Remove outliers</span></u>**

Remove outliers by calculating the Lower bound and Upper bound of the continuous columns. Any data points that exceed these bounds are considered outliers.

```python
def remove_outliers(column, bound_):
    _range = [0, 1]
    bound_range = []
    for i in _range:
        # Calculate the quantiles of the column for cid == i
        Q1 = aids_df[aids_df['cid'] == i][column].quantile(0.25)  # 25th percentile
        Q3 = aids_df[aids_df['cid'] == i][column].quantile(0.75)  # 75th percentile
        
        # Calculate the IQR (Interquartile Range)
        IQR = Q3 - Q1
        
        # Calculate the lower and upper whisker bounds
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Check if we want to find the min or max value within bounds
        if bound_ == "min":
            # Get the minimum value greater than or equal to the lower bound
            Q0 = aids_df[(aids_df['cid'] == i) & (aids_df[column] >= lower_bound)][column].min()
        elif bound_ == "max":
            # Get the maximum value less than or equal to the upper bound
            Q0 = aids_df[(aids_df['cid'] == i) & (aids_df[column] <= upper_bound)][column].max()
        
        # Append the result to the bound_range list
        bound_range.append(Q0)
    
    return bound_range

# Check results
print("Outliers of time column may fall below:", remove_outliers("time", "min"))
print("Outliers of preanti column may fall above:", remove_outliers("preanti", "max"))
print("Outliers of cd40 column may fall above:", remove_outliers("cd40", "max"))
print("Outliers of cd80 column may fall above:", remove_outliers("cd80", "max"))
print("Outliers of cd420 column may fall above:", remove_outliers("cd420", "max"))
print("Outliers of cd820 column may fall above:", remove_outliers("cd820", "max"))
```

```python
# Collect indexes of rows to drop based on outlier conditions
time_index_0 = aids_df[(aids_df['time'] < remove_outliers("time", "min")[0]) & (aids_df['cid'] == 0)].index
preanti_index_0 = aids_df[(aids_df['preanti'] > remove_outliers("preanti", "max")[0]) & (aids_df['cid'] == 0)].index
preanti_index_1 = aids_df[(aids_df['preanti'] > remove_outliers("preanti", "max")[0]) & (aids_df['cid'] == 1)].index

cd40_index_0 = aids_df[(aids_df['cd40'] > remove_outliers("cd40", "max")[0]) & (aids_df['cid'] == 0)].index
cd40_index_1 = aids_df[(aids_df['cd40'] > remove_outliers("cd40", "max")[0]) & (aids_df['cid'] == 1)].index

cd420_index_0 = aids_df[(aids_df['cd420'] > remove_outliers("cd420", "max")[0]) & (aids_df['cid'] == 0)].index
cd420_index_1 = aids_df[(aids_df['cd420'] > remove_outliers("cd420", "max")[0]) & (aids_df['cid'] == 1)].index

cd80_index_0 = aids_df[(aids_df['cd80'] > remove_outliers("cd80", "max")[0]) & (aids_df['cid'] == 0)].index
cd80_index_1 = aids_df[(aids_df['cd80'] > remove_outliers("cd80", "max")[0]) & (aids_df['cid'] == 1)].index

cd820_index_0 = aids_df[(aids_df['cd820'] > remove_outliers("cd820", "max")[0]) & (aids_df['cid'] == 0)].index
cd820_index_1 = aids_df[(aids_df['cd820'] > remove_outliers("cd820", "max")[0]) & (aids_df['cid'] == 1)].index

age_index_0 = aids_df[(aids_df['age'] > remove_outliers("age", "max")[0]) & (aids_df['cid'] == 0)].index
age_index_1 = aids_df[(aids_df['age'] > remove_outliers("age", "max")[0]) & (aids_df['cid'] == 1)].index

weight_index_0 = aids_df[(aids_df['wtkg'] > remove_outliers("wtkg", "max")[0]) & (aids_df['cid'] == 0)].index
weight_index_1 = aids_df[(aids_df['wtkg'] > remove_outliers("wtkg", "max")[0]) & (aids_df['cid'] == 1)].index

weight_index_0min = aids_df[(aids_df['wtkg'] < remove_outliers("wtkg", "min")[0]) & (aids_df['cid'] == 0)].index
weight_index_1min = aids_df[(aids_df['wtkg'] < remove_outliers("wtkg", "min")[0]) & (aids_df['cid'] == 1)].index

# Extend rows_to_drop with all the indexes
rows_to_drop = np.concatenate([time_index_0, preanti_index_0, preanti_index_1, cd40_index_0, cd40_index_1,
                               cd420_index_0, cd420_index_1, cd80_index_0, cd80_index_1, cd820_index_0, cd820_index_1,
                              age_index_0, age_index_1, weight_index_0, weight_index_1, weight_index_0min, weight_index_1min])

# Convert the result back to a pandas Index (if needed) and print
rows_to_drop_index = pd.Index(rows_to_drop)
aids_df = aids_df.drop(rows_to_drop_index.unique())
aids_df
```

### 2.2. Feature Engineering

**<u><span style="color:gray;">Add column: CD4:CD8</span></u>**

In HIV/AIDS research, the CD4:CD8 ratio is a key metric of interest. To streamline the model and reduce data dimensionality while preserving the study's significance, it can be helpful to combine the CD4 and CD8 columns to create this ratio.

Additionally, incorporating the change in the CD4:CD8 ratio before and after 20 days of treatment can provide valuable insights into the level of change in the indicator, further simplifying the model.

```python
# Divide cd4 to cd8 columns and compute differences between cd40:cd80 and cd420:cd820
aids_df['cd40/cd80'] = aids_df['cd40']/aids_df['cd80']
aids_df['cd420/cd820'] = aids_df['cd420']/aids_df['cd820']
aids_df['ratio_differences'] = aids_df['cd420/cd820'] - aids_df['cd40/cd80']
```

**<u><span style="color:gray;">Scale adjustment: Strat column</span></u>**

In the original dataset, the 'strat' variable is scored on a scale from 1 to 3 based on antiretroviral history stratification (1 = 'Antiretroviral Naive', 2 = '> 1 but <= 52 weeks of prior antiretroviral therapy', 3 = '> 52 weeks'). However, when compared to the 'str2' column, which also represents antiretroviral history, the 'naive' status is coded as 0, and 'experienced' as 1. This creates an inconsistency in the scoring system because assigning a score of 1 to 'naive', which means no prior treatment, leads to misleading results.

Therefore, in this step, the scale of the 'strat' variable will be adjusted from 1-3 to 0-2 to ensure consistency with the coding in the 'str2' column. This adjustment removes the inconsistency while maintaining the sequential meaning of treatment duration, as the scale still reflects the patient's treatment timeline.

```python
# Replace the value 1 with 0 in the 'strat' column
aids_df['strat'] = aids_df['strat'].replace(1, 0)

# modify values in the 'strat' column
aids_df['strat'] = aids_df['strat'].apply(lambda x: 1 if x == 2 else (2 if x == 3 else x))
aids_df['strat']
```

**<u><span style="color:gray;">One-hot Coding: Treatment column</span></u>**

Conversely, the 'trt' column contains four values (0, 1, 2, 3) representing four treatment methods: zdv0, zdv+ddl, zdv+zal, and ddl. The use of label encoding in this column makes it challenging to directly assess the impact of each specific treatment on the classification results. Furthermore, the 0-3 scale doesn't carry meaningful significance in terms of evaluating treatment efficacy. For example, the score of 3 does not imply that ddl is more effective than the other treatments. In fact, the ranking could be misleading because the scale does not reflect a true order of effectiveness.

As a result, in this step, the 'trt' column will be split into four separate columns, each representing one treatment method. This approach allows for clearer analysis and better interpretation of the individual effects of each treatment on the outcomes.

```python
# Create a new column 'zdv0' where the value is 1 if 'trt' is 0, otherwise 0
aids_df['zdv0'] = aids_df['trt'].apply(lambda x: 1 if x == 0 else 0)

# Create a new column 'zdv_ddl' where the value is 1 if 'trt' is 1, otherwise 0
aids_df['zdv_ddl'] = aids_df['trt'].apply(lambda x: 1 if x == 1 else 0)

# Create a new column 'zdv_zal' where the value is 1 if 'trt' is 2, otherwise 0
aids_df['zdv_zal'] = aids_df['trt'].apply(lambda x: 1 if x == 2 else 0)

# Create a new column 'ddl' where the value is 1 if 'trt' is 3, otherwise 0
aids_df['ddl'] = aids_df['trt'].apply(lambda x: 1 if x == 3 else 0)

# Display the updated dataframe
aids_df
```

**<u><span style="color:gray;">Split data & Training functions</span></u>**

Since the dataset is imbalanced, the StratifiedKFold method should be applied to split the data for training the model. The reason for using StratifiedKFold is that it ensures the samples are divided into strata with the class distribution ratio matching the overall class distribution, which is 73:26. 

Moreover, since the dataset is neither too large nor too small, using K-Fold is not computationally expensive, and it allows for more reliable testing by providing higher stability than the typical train-test split approach.

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Function to calculate the evaluation metrics for a given model
def evaluation_metrics(model, X_test, y_test):
    prediction = model.predict(X_test)
    
    accuracy_ = accuracy_score(y_test, prediction)
    precision_ = precision_score(y_test, prediction)
    recall_ = recall_score(y_test, prediction)
    f1_ = f1_score(y_test, prediction)
    
    return accuracy_, precision_, recall_, f1_
    
# Initialize Stratified K-Fold with 3 splits
foldLoop = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
## total rows = around 1700 so each fold equals to more than 500 rows -> adequate amount

# Function to train and call evaluation function
def model_training(model, X, y):
    accuracy_list = []
    precision_list = []
    recall_list = []
    f1_list = []

    # Loop through each fold
    for fold, (train_index, test_index) in enumerate(foldLoop.split(X, y)):
        X_train = X.iloc[train_index]
        X_test = X.iloc[test_index]
        y_train = y.iloc[train_index]
        y_test = y.iloc[test_index]

        model.fit(X_train, y_train)
        accuracy_, precision_, recall_, f1_ = evaluation_metrics(model, X_test, y_test)
        
        accuracy_list.append(accuracy_)
        precision_list.append(precision_)
        recall_list.append(recall_)
        f1_list.append(f1_)
        
    # Calculate average metrics across all folds
    avg_accuracy = sum(accuracy_list) / len(accuracy_list)
    avg_precision = sum(precision_list) / len(precision_list)
    avg_recall = sum(recall_list) / len(recall_list)
    avg_f1 = sum(f1_list) / len(f1_list)

    return avg_accuracy, avg_precision, avg_recall, avg_f1

# Function to print the evaluation results
def print_results(accuracy, precision, recall, f1):
    print("Best accuracy:", accuracy)
    print("Best precision:", precision)
    print("Best recall:", recall)
    print("Best f1:", f1)
```

```python
# Function to create empty dictionaries to store metrics for model evaluation 
# when looping through different parameter values (e.g., KNN neighbors or polynomial kernel degree).
def create_empty_dict():
    accuracy_dict = {}
    precision_dict = {}
    recall_dict = {}
    f1_dict = {}
    return accuracy_dict, precision_dict, recall_dict, f1_dict

# Function to find the best performance for each metric and return the key (parameter) 
# and value for the best result. The key could be 'k' for KNN or 'degree' for polynomial kernel.
def best_metrics(a_dict, p_dict, r_dict, f1_dict):
    
    best_accuracy_degree = max(a_dict, key=a_dict.get)
    best_precision_degree = max(p_dict, key=p_dict.get)
    best_recall_degree = max(r_dict, key=r_dict.get)
    best_f1_degree = max(f1_dict, key=f1_dict.get)

    best_accuracy_value = a_dict[best_accuracy_degree]
    best_precision_value = p_dict[best_precision_degree]
    best_recall_value = r_dict[best_recall_degree]
    best_f1_value = f1_dict[best_f1_degree]

    return (best_accuracy_degree, best_accuracy_value),(best_precision_degree, best_precision_value), (best_recall_degree, best_recall_value), (best_f1_degree, best_f1_value)

# Function to plot the change during loop
def plotting_loop_results(key_name, key_range, accuracy, precision, recall, f1):
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.plot(key_range, list(accuracy.values()), label='Accuracy', marker='o', linestyle='-', color='blue')
    ax.plot(key_range, list(precision.values()), label='Precision', marker='o', linestyle='-', color='green')
    ax.plot(key_range, list(recall.values()), label='Recall', marker='o', linestyle='-', color='red')
    ax.plot(key_range, list(f1.values()), label='F1 Score', marker='o', linestyle='-', color='purple')
    
    ax.set_xlabel(key_name)
    ax.set_ylabel('Metric Value')
    ax.legend()

    return fig
```

### 3.1. Logistics Regression

Parameters:
* 'class_weight="balanced"': By using "balanced", the algorithm automatically adjusts the weights for each class, giving more importance to the minority class, which helps in improving the model's ability to predict minority class instances.
* 'max_iter=200': The data does not have a complex decision boundary or too many data points, so there is no need to set the value too high at the expense of computation.

```python
from sklearn.linear_model import LogisticRegression

#logistic model
log_model = LogisticRegression(class_weight="balanced", random_state=42, max_iter=200)

#get evaluation metrics
log_accuracy, log_precision, log_recall, log_f1 = model_training(log_model, X, y)
print_results(log_accuracy, log_precision, log_recall, log_f1)
```

### 3.2. Support Vector Machine (SVM)

**<u><span style="color:gray;">RBF Kernel SVM</span></u>**

Parameters: 

* 'class_weight="balanced"': By using "balanced", the algorithm automatically adjusts the weights for each class, giving more importance to the minority class, which helps in improving the model's ability to predict minority class instances.
* gamma='scale':  Since the two classes in this dataset are not clearly separated, setting gamma too high or allowing it to be set automatically could lead to overfitting, where the model fits too closely to the noise in the training data. Using gamma="scale" helps strike a balance between underfitting and overfitting by scaling gamma according to the data's characteristics.

```python
from sklearn.svm import SVC

#rbf model
rbf_model = SVC(kernel="rbf", gamma='scale', class_weight = "balanced")

#get evaluation metrics
rbf_accuracy, rbf_precision, rbf_recall, rbf_f1 = model_training(rbf_model, X, y)

print_results(rbf_accuracy, rbf_precision, rbf_recall, rbf_f1)
```

**<u><span style="color:gray;">Polynomial Kernel SVM</span></u>**

Parameters:

* 'class_weight="balanced"': By using "balanced", the algorithm automatically adjusts the weights for each class, giving more importance to the minority class, which helps in improving the model's ability to predict minority class instances.
* gamma='scale':  Since the two classes in this dataset are not clearly separated, setting gamma too high or allowing it to be set automatically could lead to overfitting, where the model fits too closely to the noise in the training data. Using gamma="scale" helps strike a balance between underfitting and overfitting by scaling gamma according to the data's characteristics.
* degree=[2-10]: The degree of the polynomial kernel is tested over a range of values (2 to 5) because the dataset is not highly non-linear, and the classes are not perfectly separable. Using a high degree could lead to overfitting, where the model becomes too complex and fits the noise in the training data. By looping through different degrees, the model can find the optimal polynomial degree that results in the best average performance, preventing unnecessary complexity while still capturing the patterns in the data.

```python
from sklearn.svm import SVC

degree_range = [2,3,4,5,6,7,8,9,10]

#create dictionaries to store evaluation metrics
poly_accuracy_dict, poly_precision_dict, poly_recall_dict, poly_f1_dict = create_empty_dict()

#loop poly svm model through degree range
for i in degree_range:
    poly_model = SVC(kernel="poly", degree=i, gamma='scale',class_weight="balanced")
    poly_accuracy, poly_precision, poly_recall, poly_f1 = model_training(poly_model, X, y)
    
    poly_accuracy_dict[i] = poly_accuracy
    poly_precision_dict[i] = poly_precision
    poly_recall_dict[i] = poly_recall
    poly_f1_dict[i] = poly_f1

#get the best average metrics with corresponding degree
poly_best_accuracy, poly_best_precision, poly_best_recall, poly_best_f1 = best_metrics(poly_accuracy_dict, poly_precision_dict, poly_recall_dict, poly_f1_dict)

print_results(poly_best_accuracy, poly_best_precision, poly_best_recall, poly_best_f1)
chart2 = plotting_loop_results("poly_degree",degree_range, poly_accuracy_dict, poly_precision_dict, poly_recall_dict, poly_f1_dict)
```

### 3.3. K-Nearest Neighbors (KNN)

Parameter:
* k = [2 - 32, 2 steps]: Since the dataset is not large, the decision boundary is not clearly defined, and there are likely outliers, choosing a small value for k can lead to overfitting and be heavily influenced by outliers. Therefore, a range of k from 2 to 32 is selected, and looping through this range provides a better overall view of the results to identify the optimal value.
* p =1 means using the Manhattan distance instead of the default Euclidean distance. This is because the features do not have a strong or linear correlation with each other. By using Manhattan distance, the model calculates the difference between two points independently on each axis, without considering the other axes. Furthermore, since the data doesn't have a clear separable boundary, using Manhattan distance helps the model identify complex relationships between points without relying on a single straight-line boundary, allowing it to be more flexible and nuanced.

```python
from sklearn.neighbors import KNeighborsClassifier

k_range = range(2,32,2)

#create dictionaries to store evaluation metrics
knn_accuracy_dict, knn_precision_dict, knn_recall_dict, knn_f1_dict = create_empty_dict()

#loop knn model through k range
for i in k_range:
    knn = KNeighborsClassifier(n_neighbors=i, p=1)
    knn_accuracy, knn_precision, knn_recall, knn_f1 = model_training(knn, X, y)

    knn_accuracy_dict[i] = knn_accuracy
    knn_precision_dict[i] = knn_precision
    knn_recall_dict[i] = knn_recall
    knn_f1_dict[i] = knn_f1

#get the best average metrics with corresponding k
knn_best_accuracy, knn_best_precision, knn_best_recall, knn_best_f1 = best_metrics(knn_accuracy_dict, knn_precision_dict, knn_recall_dict, knn_f1_dict)

print_results(knn_best_accuracy, knn_best_precision, knn_best_recall, knn_best_f1)
chart3 = plotting_loop_results("k neighbours", k_range, knn_accuracy_dict, knn_precision_dict, knn_recall_dict, knn_f1_dict)
```

### 3.3. Decision Tree

Parameters:

* criterion="entropy": For imbalanced datasets, optimal splits are necessary to ensure that the minority class is not neglected. Entropy helps identify the best splits by minimizing uncertainty (impurity), making it a good choice for datasets with uneven distributions.
* class_weight="balanced": This parameter helps adjust the weights between classes, particularly when one class (e.g., the minority class) appears less frequently in the data. By using "balanced", it ensures that the model does not bias the majority class, helping to improve predictions for the minority class.
* max_depth=range(1, 20): This range is chosen to evaluate how the tree depth impacts the performance on an imbalanced dataset. Typically, for imbalanced problems, the tree should not be too deep to avoid overfitting while maintaining a good coverage. 
* min_samples_split=10: With imbalanced data, excessively small splits can lead to the model learning from noise or outliers. Setting min_samples_split=10 ensures that a node must have at least 10 samples before it can be split, helping to prevent overfitting and making the model more robust to outliers.
* min_samples_leaf=5: This parameter defines the minimum number of samples that each leaf node must contain. If a leaf node contains too few samples, the model may fit overly specific patterns from features that are not important, leading to overfitting. Setting min_samples_leaf=5 ensures each leaf node is based on a sufficient amount of data, making the model more generalizable.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import make_classification

depth_range = range (1,20)

#create dictionaries to store evaluation metrics
tree_accuracy_dict, tree_precision_dict, tree_recall_dict, tree_f1_dict = create_empty_dict()

#loop decision tree model through depth
for i in depth_range:
    tree_model = DecisionTreeClassifier(criterion="entropy", class_weight="balanced", max_depth=i, min_samples_split=10, min_samples_leaf=5)
    tree_accuracy, tree_precision, tree_recall, tree_f1 = model_training(tree_model, X, y)

    tree_accuracy_dict[i] = tree_accuracy
    tree_precision_dict[i] = tree_precision
    tree_recall_dict[i] = tree_recall
    tree_f1_dict[i] = tree_f1

#get the best average metrics with corresponding depth
tree_best_accuracy, tree_best_precision, tree_best_recall, tree_best_f1 = best_metrics(tree_accuracy_dict, tree_precision_dict, tree_recall_dict, tree_f1_dict)

print_results(tree_best_accuracy, tree_best_precision, tree_best_recall, tree_best_f1)

chart4 = plotting_loop_results("max depth", depth_range, tree_accuracy_dict, tree_precision_dict, tree_recall_dict, tree_f1_dict)
```
